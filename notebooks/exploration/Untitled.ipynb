```python
# ── Basic function ────────────────────────────────────────────────
def calculate_tip_percentage(fare: float, tip: float) -> float:
    '''
    Calculate tip as percentage of fare.

    Args:
        fare: Base fare amount in dollars
        tip: Tip amount in dollars

    Returns:
        Tip percentage (0-100)
    '''
    if fare <= 0:
        return 0.0
    return round((tip / fare) * 100, 2)
```

```python
class trip:

    def __init__(self, fare:float, distance: float):
        self.fare = fare
        self.distance = distance

trip1 = trip(12.5,2.3)
trip2 = trip(fare=45,distance=8.7)


# Access attributes
print(trip1.fare)        # 12.5
print(trip2.distance)    # 8.7
print(trip1.fare)        # 12.5 — trip1 not affected by trip2
```

```python
class RideShareCompany:
    # CLASS ATTRIBUTE - shared by ALL instances
    commission_rate = 0.25  # 25% commission
    total_rides = 0
    
    def __init__(self, driver_name: str):
        # INSTANCE ATTRIBUTE - unique to each instance
        self.driver_name = driver_name
        self.earnings = 0
    
    def complete_ride(self, fare: float):
        company_cut = fare * RideShareCompany.commission_rate
        driver_cut = fare - company_cut
        
        self.earnings += driver_cut  # instance attribute
        RideShareCompany.total_rides += 1  # class attribute

# Create drivers
driver_a = RideShareCompany("Alice")
driver_b = RideShareCompany("Bob")

driver_a.complete_ride(100)
driver_b.complete_ride(200)

# Instance attributes differ
print(driver_a.earnings)  # 75.0 (Alice's earnings)
print(driver_b.earnings)  # 150.0 (Bob's earnings)

# Class attribute is shared
print(RideShareCompany.total_rides)  # 2 (total across all drivers)
print(driver_a.total_rides)
print(driver_b.total_rides)
```

```python
class Temperature:
    def __init__(self, celsius: float):
        self._celsius = celsius
    
    # ═══════════════════════════════════════
    # OPTION 1: Using @property (RECOMMENDED)
    # ═══════════════════════════════════════
    
    @property
    def celsius(self) -> float:
        return self._celsius
    
    @celsius.setter
    def celsius(self, value: float):
        if value > -273.15:  # Absolute zero
            raise ValueError("Temperature below absolute zero")
        self._celsius = value
    
    @property
    def fahrenheit(self) -> float:
        """Computed on-the-fly"""
        return (self._celsius * 9/5) + 32
    
    @fahrenheit.setter
    def fahrenheit(self, value: float):
        """Convert and store as Celsius"""
        self.celsius = (value - 32) * 5/9
    
    @property
    def kelvin(self) -> float:
        """Read-only computed property"""
        return self._celsius + 273.15


# Clean, intuitive usage
temp = Temperature(30)
print(temp.celsius)      # 25 — clean access
print(temp.fahrenheit)   # 77.0 — computed automatically
print(temp.kelvin)       # 298.15 — read-only

temp.fahrenheit = 100    # Set using Fahrenheit
print(temp.celsius)      # 37.78 — automatically converted
```

```python
# ═══════════════════════════════════════
# OPTION 2: Using methods (OLD STYLE)
# ═══════════════════════════════════════

class TemperatureOldStyle:
    def __init__(self, celsius: float):
        self._celsius = celsius
    
    def get_celsius(self) -> float:
        return self._celsius
    
    def set_celsius(self, value: float):
        if value < -273.15:
            raise ValueError("Temperature below absolute zero")
        self._celsius = value
    
    def get_fahrenheit(self) -> float:
        return (self._celsius * 9/5) + 32
    
    def set_fahrenheit(self, value: float):
        self.set_celsius((value - 32) * 5/9)
    
    def get_kelvin(self) -> float:
        return self._celsius + 273.15


# Ugly, verbose usage
temp_old = TemperatureOldStyle(25)
print(temp_old.get_celsius())        # ❌ Verbose
print(temp_old.get_fahrenheit())     # ❌ Inconsistent with attributes
print(temp_old.get_kelvin())         # ❌ More typing

temp_old.set_fahrenheit(100)         # ❌ Ugly setter
print(temp_old.get_celsius())        # ❌ Can't just do temp_old.celsius
```

```python
import pandas as pd
import numpy as np

# Load NYC Taxi data
df = pd.read_parquet("yellow_tripdata_2024-01.parquet")

# ── Essential pandas operations ───────────────────────────────────

# Selecting columns
fares = df["fare_amount"]                           # single column → Series
money_cols = df[["fare_amount", "tip_amount"]]      # multiple → DataFrame

# Filtering rows
valid = df[df["fare_amount"] > 0]                   # simple filter
quality = df[
    (df["fare_amount"] > 0) &                       # AND
    (df["trip_distance"] > 0) &
    (df["passenger_count"] > 0)
]

# Creating new columns
df["trip_duration_min"] = (
    pd.to_datetime(df["tpep_dropoff_datetime"]) -
    pd.to_datetime(df["tpep_pickup_datetime"])
).dt.total_seconds() / 60

df["tip_percentage"] = (
    df["tip_amount"] / df["fare_amount"].replace(0, np.nan) * 100
).round(2)

df["speed_mph"] = (
    df["trip_distance"] / (df["trip_duration_min"] / 60)
).replace([np.inf, -np.inf], np.nan)

# Aggregations — GROUP BY equivalent
hourly_stats = (
    df
    .assign(hour=pd.to_datetime(df["tpep_pickup_datetime"]).dt.hour)
    .groupby("hour")
    .agg(
        trip_count=("fare_amount", "count"),
        avg_fare=("fare_amount", "mean"),
        total_revenue=("total_amount", "sum"),
        avg_tip_pct=("tip_percentage", "mean")
    )
    .round(2)
    .reset_index()
    .sort_values("hour")
)

print(hourly_stats.head(10))
```
